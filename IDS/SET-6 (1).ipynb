```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data={
    'X': np.random.rand(100),
    'Y': np.random.rand(100)
}

df=pd.DataFrame(data)
df.head()

plt.plot(df['X'],df['Y'])
plt.title("Line plot of Random Number")
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data={
    'X': np.random.rand(100),
    'Y': np.random.rand(100)
}

df=pd.DataFrame(data)
df.head()

plt.scatter(df['X'],df['Y'])
plt.title("Scatter plot of Random Number")
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data=np.random.rand(100)

df=pd.DataFrame(data)
df.head()

plt.hist(data,bins=15,edgecolor='black')
plt.title("Histogram of Random Number")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
