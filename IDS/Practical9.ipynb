```python
#Load California Housing data set

from sklearn.datasets import fetch_california_housing
housingData = fetch_california_housing()
print(housingData)
```

```python
#Split the data into training and tesing data set

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(housingData.data, housingData. target, test_size=0.30,random_state=35)

print("Shape of training input data: ",x_train.shape)
print("Shape of testing input data,x_test.shape")
```

```python
#Creation of Linear Regression Model
from sklearn.linear_model import LinearRegression
linearReg=LinearRegression()
```

```python
#Training Linear Regression Model with training data
linearReg.fit(x_train,y_train)
```

```python
#Testing Linear Regression model with testing data
predictResult = linearReg.predict(x_test)
print(predictResult)
```

```python
#Calculate accuracy metrics for Linear Regression model

from sklearn.metrics import mean_squared_error, r2_score

linear_mse = mean_squared_error(y_test,predictResult)

print("Mean squared error for linear regression model: ",linear_mse)

linear_r2=r2_score(y_test, predictResult)

print("R2 score for linear regression model: ", linear_r2)
```

```python
#Creation of Random Forest model
from sklearn.ensemble import RandomForestRegressor
random_Reg = RandomForestRegressor()
```

```python
#Training Random Forest Regression Model
random_Reg.fit(x_train,y_train)
```

```python
#Testing Random Forest Regression Model
randomPredicted = random_Reg.predict(x_test)
randomPredicted
```

```python
#Calculate accuracy metrics for Random Forest Regression model

from sklearn.metrics import mean_squared_error, r2_score

random_mse = mean_squared_error(y_test, randomPredicted)

print("Mean squared error for random forest regression model: ",random_mse)

random_r2=r2_score(y_test, randomPredicted)

print("R2 score for linear regression model: ",random_r2)
```
