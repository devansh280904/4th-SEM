```python
import pandas as pd
import numpy as np

exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily',
    'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df=pd.DataFrame(exam_data, index=labels)

 #Select and display the 'name'and'score'columns
print("Name and Score columns:")
print(df[['name','score']])

#Select 'score' and ‘qualify’ columns in rows1,3,5 and 6
print("\nScore and Qualify coolumns in rows 1, 3, 5 and 6:")
print(df.loc[['b','d','f','g'],['score','qualify']])

#Display the rowswhere thenumber of attempts>2
print("\nRows where the number of attempts > 2:")
print(df[df['attempts']>2])

#Display the rows where the score is missing(NaN)
print("\nRows where the score is missing:")
print(df[df['score'].isnull()])

#Display rows where the score is between 15 and 20 (inclusive)
print("\nScore between 15 and 20:")
print(df[(df['score']>=15)&(df['score']<=20)])

#Change the score in row 'd' to 11.5
print("\nChanging score in row 'd' to 11.5:")
df.at['d','score']=11.5
print(df)
     
#Create a data frame grouped by attempts and display the mean of score for each group
print("\nMean score of each group based on attempts:")
grp=df.groupby('attempts')['score'].mean()
print(grp)
```

```python
import pandas as pd

df=pd.read_csv("Missing-employees.csv")

#Display Number of rows
nrows=df.shape[0]
print("Number of rows:",nrows)

#Display Number of attributes (columns)
nattributes=df.shape[1]
print("\nNmber of attribute:",nattributes)

#Display properties of each attribute
attribute_properties=df.dtypes
print("\nProperties of each attribute:")
print(attribute_properties)

#Number of missing values for each attribute
print("\nNumber of missing values for each attribute:")
mv=df.isnull().sum()
print(mv)

#Total number of missing values
tmv=df.isnull().sum().sum()
print("\nTotal number of missing values:",tmv)

#Replace missing values with different methods
print("\nReplacing missing values:")

#with constant value
print("\nReplacing with constant value:")
df_const=df.fillna(0)
print("Displaying first few rows after replacing missing value:")
print(df_const.head())

#with mean value
print("\nReplacing with constant value:")
df_m=df.fillna(df.mean)
print("Displaying first few rows after replacing missing value:")
print(df_m.head())

#fill using backfill
print("\nFilling using backfill:")
df_bf=df.fillna(method='backfill')
print("Displaying first few rows after replacing missing value:")
print(df_bf.head())

# Drop null values
print("\nDropping null values:")
df_dropna=df.dropna()







```
