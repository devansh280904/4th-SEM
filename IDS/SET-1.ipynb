```python
# Write a program to take input from users and print if the input number is odd or even.

a=int(input("ENTER YOUR NUMBER :"))
if a%2==0:
    print(a,"Given Number is even")
else:
    print(a,"Given Number is Odd")
```

```python
# Write a program to find if the number is positive, negative or zero.

a=int(input("ENTER YOUR NUMBER :"))
if a<0:
  print(a,"NUMBER IS NEGATIVE")
elif a>0:
  print(a,"NUMBER IS POSITIVE")
else:
  print(a,"NUMBER IS 0")
```

```python
# Write a program to print all even numbers within the range provided by the user.

print("ENTER START VALUE")
n=int(input())
print("ENTER END VALUE")
b=int(input())
for i in range(n,b):
    if i%2==0:
        print(i)
```

```python
#Write a program to compute the factorial of a number given by the user. (while)

a=int(input("ENTER:"))
i=1
fact=1
while i<=a:
    fact=fact*i
    i+=1
print(fact)
```

```python
#Write a function to display digit at unit place of a given number

def unit(n):
    i=n%10
    print(i)
n=int(input("ENTER THE NUMBER:"))

unit(n)

```

```python
#Write a function to display maximum of three numbers

def m(a,b,c):

    if a>b:
        if a>c:
            print(a,"IS MAXIMUM")
    elif b>c:
        print(b,"IS MAXIMUM")
    else:
        print(c,"IS MAXIMUM")
a=int(input("ENTER THE 1st NUMBER:"))
b=int(input("ENTER THE 2nd NUMBER:"))
c=int(input("ENTER THE 3rd NUMBER:"))

m(a,b,c)
```
